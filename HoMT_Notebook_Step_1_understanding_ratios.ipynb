*   Understanding concepts and ratios in each tuning system and temperament by answering questions.




```python
#DO NOT CHANGE THE CODE HERE
import numpy as np
import IPython.display as Audio
import pandas as pd

def ask_question(question, correct_answer):
    while True:
        answer = input(question)
        if answer == correct_answer:
            print("Correct!")
            return answer
        else:
            print("Wrong. Try again!")
```

```python
# Pythagorean tuning
print("Let's complete the diatonic scale in Pythagorean tuning!")

# Step 1: G2
G2 = ask_question("Make G2 by multiplying 3 to C1 (e.g., 3/1): ", "3/1")

# Step 2: G1
G1 = ask_question("Make it one octave lower by multiplying 1/2 (e.g., 3/2): ", "3/2")

# Step 3: D2
D2 = ask_question("Make D2 by multiplying perfect fifth to G1 (3/2 * 3/2 = ?): ", "9/4")

# Step 4: D1
D1 = ask_question("Make it one octave lower by multiplying 1/2 (9/4 * 1/2 = ?): ", "9/8")

# Step 5: A1
A1 = ask_question("Make A1 by multiplying perfect fifth to D1 (9/8 * 3/2 = ?): ", "27/16")

# Step 6: E2
E2 = ask_question("Make E2 by multiplying perfect fifth to A1 (27/16 * 3/2 = ?): ", "81/32")

# Step 7: E1
E1 = ask_question("Make it one octave lower by multiplying 1/2 (81/32 * 1/2 = ?): ", "81/64")

# Step 8: B1
B1 = ask_question("Make B1 by multiplying perfect fifth to E1 (81/64 * 3/2 = ?): ", "243/128")

# Step 9: F0
F0 = ask_question("Make F0 by dividing 3/2 of C1 (eg., 2/3): ", "2/3")

# Step 10: F1
F1 = ask_question("Make it one octave higher by multiplying 2 (eg., 2/3 * 2 = ?): ", "4/3")

print("Congratulations! You completed the diatonic scale in Pythagorean tuning!")

df = pd.DataFrame({
    "C1": [1],
    "D1": ["9/8"],
    "E1": ["81/64"],
    "F1": ["4/3"],
    "G1": ["3/2"],
    "A1": ["27/16"],
    "B1": ["243/128"],
    "C2": [2]
})

styled_df = df.style.apply(lambda x: ['background-color: yellow' if x.name == 'G1' else '' for _ in x], axis=0)
styled_df
```

## 2) Just Intonation

```python
#Just Intonation

Major_third_in_PT = ask_question("What was the ratio of Major third in Pythagorean tuning?", "81/64")
print("Let's listen to this major third")
C4_base_freq = 260
major_third_freq = C4_base_freq * 81/64
t = np.linspace(0, 2, 22050 *2, False)
C4_tone = 0.5 * np.sin(2 * np.pi * C4_base_freq * t)
major_third_tone = 0.5 * np.sin(2 * np.pi * major_third_freq * t)
tone_alternating = np.concatenate([C4_tone, major_third_tone])
major_third = 0.5 * (np.sin(2 * np.pi * C4_base_freq * t) + np.sin(2 * np.pi * major_third_freq * t))
combined_tones = np.concatenate([tone_alternating, major_third])
Audio.Audio(combined_tones, rate=22050)
```

```python
print("Now listen to this. This time we'll use purer third by using simpler ratio 5/4, which is closest to 81/64 in Pythagorean tuning.")

C4_base_freq = 260
major_third_freq = C4_base_freq * 5/4
t = np.linspace(0, 2, 22050 *2, False)
C4_tone = 0.5 * np.sin(2 * np.pi * C4_base_freq * t)
major_third_tone = 0.5 * np.sin(2 * np.pi * major_third_freq * t)
tone_alternating = np.concatenate([C4_tone, major_third_tone])
major_third = 0.5 * (np.sin(2 * np.pi * C4_base_freq * t) + np.sin(2 * np.pi * major_third_freq * t))
combined_tones = np.concatenate([tone_alternating, major_third])
Audio.Audio(combined_tones, rate=22050)
```

## 3) Meantone Temperament


```python
from fractions import Fraction
print("Meantone Temperament reduces the size of the fifth to get purer major third. I'll let you know the ratios of the quartertone comma meantone. Answer the ratio of the fifth.")

syntonic_comma = (81 / 80)
G = (3 / 2) * (syntonic_comma ** (-1/4)) # Reduce by 1/10 comma
if G > 2:
    G /= 2
G_fraction = Fraction(G).limit_denominator()
print ("G:", G_fraction)

D = G_fraction * G_fraction
if D > 2:
    D /= 2
D_fraction = Fraction(D).limit_denominator()
print ("D:", D_fraction)

A = D_fraction * G_fraction
if A > 2:
    A /= 2
A_fraction = Fraction(A).limit_denominator()
print ("A:", A_fraction)

E = A_fraction * G_fraction
if E > 2:
    E /= 2
E_fraction = Fraction(E).limit_denominator()
print ("E:", E_fraction)

B = E_fraction * G_fraction
if B > 2:
    B /= 2
B_fraction = Fraction(B).limit_denominator()
print ("B:", B_fraction)

F = 1 / G  # Inverse of the fifth
if F < 1:  # If F is less than one octave below
    F *= 2
F_fraction = Fraction(F).limit_denominator()
print("F:", F_fraction)

M_G = ask_question("What is the ratio of G in Meantone temperament?", "1242261/830750")

print("Scroll up and compare with the highlighted ratio (decimal) of E4 in just intonation. Did G4 get smaller in Meantone tuning?" )
df = pd.DataFrame({
    "C4": [1],
    "D4": [930249/832040],
    "E4": [5/4],
    "F4": [955341/714284],
    "G4": [1242261/830750],
    "A4": [539966/322975],
    "B4": [1242261/664600],
    "C5": [2/1]
})

styled_df = df.style.apply(lambda x: ['background-color: yellow' if x.name == 'G4' else '' for _ in x], axis=0)
styled_df


```
